#### 도미 데이터 준비하기
여러 개의 도미 생선 데이터를 가지고 어떤 생선이 도미인지를 구분할 기준을 구함  

- 이진 분류 : 머신러닝에서 여러 개의 종류 중 하나를 구별해 내는 문제를 분류라고 함
- 특히 2개의 클래스 중 하나를 고르는 묹를 이진 분류라고 함
- 파이썬 프로그램의 클래스와는 다른 개념이니 혼동하면 안 됨

```python
# 첫번째 특성(Feature)
# 생선 길이
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
# 두번째 특성(Feature)
# 생선 무게
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

- 두 특성을 그래프로 표현하여 확인함  
- 그래프는 matplotlib 패키지를 사용함
- 산점도 : x, y 축으로 이뤄진 좌표계에 두 변수의 관계를 표현함

```python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

- 도미와 비교를 위한 빙어 데이터 업로드

```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

- K-Nearest Neighbors 알고리즘을 사용해 도미와 빙어 데이터 구분
- 해당 알고리즘 사용을 위하여 도미와 빙어의 데이터를 합침

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

- 사이킷런 : 머신러닝 패키지
- 2차원 리스트가 필요함
- 2차원 리스트를 만들기 위하여 'zip()' 함수를 사용함

```python
fish_data = [[l, w] for l, w in zip(length, weight)]
print(fish_data)
```

- 도미의 데이터와 빙어의 데이터 범위가 어디인지 알려줘야 함
- 데이터를 주어도 어떤 생선인지 모르기 때문에 미리 알려주어야 함
- 도미는 '1', 빙어는 '0'으로 표기함
- 도미 데이터는 총 35개이고, 빙어 데이터는 총 14개이므로 해당 갯수만큼의 배열을 생성함

```python
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

- 객체에 fish_data와 fish_target을 전달하여 도미를 찾기 위한 기준을 학습 시킴
- 머신러닝에서는 '훈련' 이라고 부름

```python
kn.fit(fish_data, fish_target)
kn.score(fish_data, fish_target)
```

- fit() 메서드는 주어진 데이터로 알고리즘을 훈련시킴
- 객체 kn이 얼마나 잘 훈련되었는지 평가하려면 score() 메서드를 사용함
- predict는 새로운 데이터의 정답을 예측함
- 도미와 비슷한 길이와 무게를 넣으면 도미의 값인 1을 반환함

```python
kn.predict([[30, 600]])
```

- _fit_X 속성에는 전달한 fish_data가 저장되어 있음
- _y 속성에는 fish_target이 저장되어 있음

```python
print(kn._fit_X)
print(kn._y)
```

```python
# 가까운 데이터 몇 개를 참고할까?
# 정하기 나름임
# 참고 데이터를 49개로 한 kn49 모델
# 가장 가까운 데이터 49개를 사용하는 K-Nearest Neighbors를 이용하면 데이터 49개 중 도미가 35개로 다수를 차지하므로,
# 어떤 데이터를 넣어도 무조건 도미로 예측할 것임
kn49 = KNeighborsClassifier(n_neighbors=49)
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
# 49개의 데이터 중 35개가 도미의 데이터이므로 정확도를 계산하면, 35/49가 됨
print(35/49)
```
